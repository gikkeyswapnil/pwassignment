# Q1. What is Flask Framework? What are the advantages of Flask Framework?
## Ans:
### Flask Framework: Flask is a popular Python web framework that provides a lightweight and flexible approach to building web applications. It was created by Armin Ronacher in 2010 and is known for its simplicity, modularity, and scalability. Flask is classified as a "micro-framework" because it doesn't require particular tools or libraries to run, and it is not tightly coupled to a particular database or other technology stack.
### The advantages of using Flask :

1. Lightweight and flexible: Flask is designed to be lightweight and flexible, which makes it easy to use and customize according to your project's specific needs.

1. Easy to get started: Flask has a simple and intuitive API, and its documentation is comprehensive and easy to follow, making it easy for beginners to get started.

1. Modular: Flask allows you to use only the components you need and integrate them with other technologies, such as databases, front-end frameworks, or authentication systems.

1. Extensible: Flask has a large and active community that has created numerous plugins and extensions that add functionality to the framework, such as authentication, database integration, and API development.

1. Fast and efficient: Flask is known for its fast performance and efficient memory usage, which makes it ideal for building high-performance web applications.

1. Open source: Flask is an open-source framework, which means it is free to use and can be customized according to your specific needs

# Q3. What is App routing in Flask? Why do we use app routes?
## Ans:
### App  Routing : In Flask, app routing refers to the process of mapping URL paths to the appropriate view function that should handle the request. App routing is a fundamental part of building a Flask application, as it allows you to define the behavior of your application and how it should respond to different requests.Flask provides a simple and intuitive way to define app routes using the @app.route decorator. The @app.route decorator is used to associate a URL with a view function. When a request is made to the URL that matches the route, Flask calls the associated view function, which returns a response to the client.

### We use app routes in Flask for several reasons, including:

1. Defining the behavior of our application: App routes allow us to define the behavior of our application and how it should respond to different requests.

1. Organizing our code: App routes provide a way to organize our code and separate different parts of our application into different views or functions.

1. Creating a RESTful API: App routes can be used to create a RESTful API that allows other applications to interact with our application and exchange data.

1. Building complex applications: App routes are essential when building complex applications that require multiple views and functionality.

# Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.
## Ans: In Flask, the url_for() function is used for URL building. This function generates a URL for the specified endpoint, using the URL rules that are defined in the Flask application.The url_for() function takes the endpoint name as its first argument and any number of keyword arguments representing the values of the dynamic parts of the URL rule.

```python
#code
import flask from Flask ,url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'Hello, world!'

@app.route('/user/<username>')
def profile(username):
    return f'Hello, {username}!'

with app.test_request_context():
    print(url_for('index'))
    print(url_for('profile', username='aniruddha'))
```
