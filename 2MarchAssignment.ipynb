## Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.
### Ans:
* Matplotlib is a popular data visualization library for Python that allows users to create a variety of static, interactive, and animated visualizations. It provides a range of tools for constructing graphs, charts, histograms, scatterplots, and other types of visualizations.Matplotlib is used because of its ability to produce high-quality, publication-ready plots with a minimal amount of code

* Five plots that can be plotted using the Pyplot module of Matplotlib are:

1. Line Plot: A line plot is a graph that shows the relationship between two variables on a line. It is used to visualize trends or patterns over time.

1. Scatter Plot: A scatter plot is a graph that shows the relationship between two variables. It is used to identify correlations or clusters in the data.

1. Histogram: A histogram is a graph that shows the distribution of a dataset. It is used to visualize the frequency of occurrence of a variable.

1. Bar Plot: A bar plot is a graph that shows the comparison between different categories. It is used to compare the values of a variable across different categories.

1. Pie Chart: A pie chart is a graph that shows the proportion of each category in a dataset. It is used to visualize the relative size of different categories.

## Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

### Ans:
* Scatter Plot : A scatter plot is a type of graph that displays the relationship between two continuous variables. It uses dots or markers to represent individual data points, and their position on the plot corresponds to the value of the two variables.Each point on the scatter plot represents a single observation, where the x-axis represents one variable and the y-axis represents the other variable. By plotting these data points on the graph, we can easily identify any patterns, trends, or relationships between the two variables.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.title("Example of Scatter Plot")
```

## Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
### Ans : The subplot() function is used to create multiple plots within a single figure. It allows us to arrange multiple plots in a grid-like pattern, where each plot represents a different subset of the data or a different aspect of the analysis.

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# subplots
fig, axs = plt.subplots(nrows=2, ncols=2)

# Plot line 1
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

# Plot line 2
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

# Plot line 3
axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

# Plot line 4
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

# Set the layout of the subplots
plt.tight_layout()
plt.show()
```

## Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
### Ans: A bar plot, also known as a bar chart, is a type of graph that uses bars to represent categorical data. Each bar in the plot corresponds to a particular category, and the length or height of the bar represents the value of a numerical variable associated with that category.Bar plots are commonly used to compare the values of different categories, and to show how these values change over time or across different groups. They are particularly useful when the data is discrete, non-continuous or non-linear.Bar plots can also be used to display the frequency or count of different categories, such as the number of people in each age group, or the number of cars sold by different manufacturers.

```python
import numpy as np
import matplotlib.pyplot as plt
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.title("Bar plot")
plt.xlabel("company name")
plt.ylabel("profit ")
plt.show()
```

## Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
### Ans: A box plot, also known as a box and whisker plot, is a type of graph that displays the distribution of a dataset. It uses a rectangular box to represent the range of the data, and two whiskers to represent the spread of the data.The box in the plot represents the interquartile range (IQR), which is the range of the middle 50% of the data. The whiskers represent the lowest and highest values in the data that fall within 1.5 times the IQR from the upper and lower quartiles, respectively. Any data points that fall outside this range are considered outliers and are shown as individual points on the plot.Box plots are commonly used to visualize and compare the distribution of data across different groups or categories. They are particularly useful when the data is non-normal or skewed, and when there are outliers present in the dataset. Box plots allow us to identify the central tendency, variability, and skewness of the data, and to detect any potential outliers or unusual observations.Box plots are widely used in many fields, including statistics, finance, economics, and biology. They are a powerful tool for summarizing and communicating complex data in a simple and intuitive wa

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate the data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a figure and axis
fig, ax = plt.subplots()

# Create the box plot
data = [box1, box2]
ax.boxplot(data)

# Set the labels and title
ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_ylabel('Value')
ax.set_title('Box Plot of Two Distributions')

# Show the plot
plt.show()
```
